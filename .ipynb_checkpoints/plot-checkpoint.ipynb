```python
import matplotlib.pyplot as plt
import json
import random

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
```

```python
ticks = -1
nodes = {}

with open('./data/result.json') as f:
    result = json.load(f)
    ticks = result["ticks"]
    for key in result["nodes"]:
        nodes[key] = result["nodes"][key]
```

```python


fig = plt.figure(1, figsize=(20, 10))

plt.subplot(1,1,1)
for key in nodes:
    color = "-" + random.choice("rbgy")
    plt.plot([i for i in range(1, ticks + 1)], nodes[key], color, label=key)
plt.title("Нагрузка на вычислительном кластере при балансировке WeightedRoundRobin")
plt.xlabel("Момент времени")
plt.ylabel("Количество выполняемых задач")
plt.legend(loc="upper left")
plt.grid(True)
```
