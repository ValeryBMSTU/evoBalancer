```python
import matplotlib.pyplot as plt
import json
import random

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 22}

plt.rc('font', **font)
```

```python
ticks = -1
nodes = {}

with open('./data/results/age0_agent0.json') as f:
    result = json.load(f)
    ticks = result["ticks"]
    for key in result["nodes"]:
        nodes[key] = result["nodes"][key]
```

```python


fig = plt.figure(1, figsize=(20, 10))

plt.subplot(1,1,1)
colorIndex = 0
colors = "rbgyck"
for key in nodes:
    plt.plot([i for i in range(1, ticks + 1)], nodes[key], colors[colorIndex], label=key)
    colorIndex = colorIndex + 1
plt.title("Нагрузка на вычислительном кластере при балансировке WeightedRoundRobin")
plt.xlabel("Момент времени")
plt.ylabel("Количество выполняемых задач")
plt.legend(loc="upper left")
plt.grid(True)
```

```python
nodes = {}
scores = []

with open('./data/results/evo_result.json') as f:
    result = json.load(f)
    scores = result["scores"]
    for key in result["nodes"]:
        nodes[key] = result["nodes"][key]
```

```python


fig = plt.figure(1, figsize=(20, 10))

plt.subplot(1,1,1)
colorIndex = 0
colors = "rbgyck"
for key in nodes:
    plt.plot([i for i in range(1, len(nodes[key])+1)], nodes[key], colors[colorIndex], label=key)
    colorIndex = colorIndex + 1
plt.title("Эволюционный процесс")
plt.xlabel("Эпоха")
plt.ylabel("Веса")
plt.legend(loc="upper left")
plt.grid(True)
```
